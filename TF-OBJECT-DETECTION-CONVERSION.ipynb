```python
import tensorflow as tf

# Load the SavedModel
model = tf.saved_model.load("/content/gdrive/MyDrive/saved_model")

# Get the serving_default function
infer = model.signatures['serving_default']

# Prepare a sample input tensor. Adjust the shape and data type according to your model's needs.
# Example for an input tensor of shape (1, 300, 300, 3) and type float32
input_tensor = tf.random.normal([1, 300, 300, 3])

# Run inference
output = infer(input=input_tensor)

# Output is typically a dictionary with output tensor names as keys and Tensor values
print(output)
print("Input details:")
for input_name, input_value in infer.structured_input_signature[1].items():
    print(f"Name: {input_name}, Shape: {input_value.shape}, Type: {input_value.dtype}")

print("\nOutput details:")
for output_name, output_value in infer.structured_outputs.items():
    print(f"Name: {output_name}, Shape: {output_value.shape}, Type: {output_value.dtype}")

# Let's assume you have the concrete function and want to save it as a frozen graph
from tensorflow.python.framework import convert_to_constants

# Convert the concrete function to a fully frozen graph
frozen_func = convert_to_constants.convert_variables_to_constants_v2(concrete_func)

# Print the frozen graph's ops, to validate
for op in frozen_func.graph.get_operations():
    print(op.name)

# Save the frozen graph to a file
tf.io.write_graph(graph_or_graph_def=frozen_func.graph.as_graph_def(),
                  logdir="/content/gdrive/MyDrive/saved_model",
                  name="frozen_graph.pb",
                  as_text=False)

```

```python
import tensorflow as tf

# Load the frozen graph
graph_def = tf.compat.v1.GraphDef()
with tf.io.gfile.GFile('/content/gdrive/MyDrive/saved_model/frozen_graph.pb', 'rb') as f:
    graph_def.ParseFromString(f.read())

# Import the graph_def into a new Graph and returns it
with tf.Graph().as_default() as graph:
    tf.import_graph_def(graph_def, name='')

# Print the operation names to find the output node
for op in graph.get_operations():
    print(op.name)
```

```python
import tensorflow as tf
from tensorflow.python.framework.convert_to_constants import convert_variables_to_constants_v2

def load_and_freeze_model(saved_model_dir, output_graph_filename):
    # Load the TensorFlow SavedModel
    model = tf.saved_model.load(saved_model_dir)
    concrete_func = model.signatures['serving_default']

    # Concrete function already defines inputs and outputs, so use directly
    frozen_func = convert_variables_to_constants_v2(concrete_func)
    frozen_graph_def = frozen_func.graph.as_graph_def()

    # Print the layers of the frozen model
    layers = [op.name for op in frozen_func.graph.get_operations()]
    print("-" * 50)
    print("Frozen model layers: ")
    for layer in layers:
        print(layer)
    print("-" * 50)

    # Print the inputs and outputs of the frozen model
    print("Frozen model inputs: ", frozen_func.inputs)
    print("Frozen model outputs: ", frozen_func.outputs)

    # Save the frozen graph to a file
    tf.io.write_graph(frozen_graph_def, '.', output_graph_filename, as_text=False)
    print(f"Frozen model has been saved as {output_graph_filename}")

# Specify the SavedModel directory and output filename
saved_model_dir = '/content/gdrive/MyDrive/saved_model'
output_graph_filename = '/content/gdrive/MyDrive/saved_model/frozen_graph.pb'

# Call the function to load and freeze the model
load_and_freeze_model(saved_model_dir, output_graph_filename)
```
